```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
realCIR = np.load('./data 2d/real_CIR.npy')
imagCIR = np.load('./data 2d/imag_CIR.npy')
GT = np.load('./data 2d/GT.npy')
```

```python
print(realCIR.shape,imagCIR.shape)
```

```python
data = np.stack((realCIR,imagCIR),axis=-1)
```

```python
data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, GT, test_size=0.2, shuffle=True, stratify=GT)
```

```python
model = tf.keras.models.Sequential([
            tf.keras.layers.Flatten(input_shape=(128,2)),
            tf.keras.layers.Dense(200,activation=tf.nn.relu),
            tf.keras.layers.BatchNormalization(axis=-1),
            tf.keras.layers.Dropout(0.2),
            tf.keras.layers.Dense(400,activation=tf.nn.relu),
            tf.keras.layers.BatchNormalization(axis=-1),
            tf.keras.layers.Dropout(0.2),
            tf.keras.layers.Dense(200,activation=tf.nn.relu),
            tf.keras.layers.BatchNormalization(axis=-1),
            tf.keras.layers.Dropout(0.2),
            tf.keras.layers.Dense(6,activation='softmax')
])
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler
```

```python
def step_decay(epoch):
    start = 0.001
    drop = 0.1
    epochs_drop = 10
    lr = start * (drop ** np.floor((epoch)/epochs_drop))
    return lr
```

```python
lr_scheduler = LearningRateScheduler(step_decay, verbose=1)
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100,callbacks=[lr_scheduler])
```

```python
model.evaluate(X_test, y_test)
```
